```python
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\HP\\OneDrive\\Documents\\datasets\\Titanic-Dataset.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data[['Age']] = data[['Age']].fillna(data[['Age']].median())
```

```python
data.columns = data.columns.str.strip()   # remove spaces from column names
print(data.columns)
```

```python
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])

# data['Embarked'] = le.fit_transform(data['Embarked']) 
data = pd.get_dummies(data, columns=['Embarked'], drop_first=True)
```

```python
data = data.drop(['PassengerId', 'Ticket', 'Cabin'], axis=1)
data['FamilySize'] = data['SibSp'] + data['Parch']+1

data.drop(['SibSp', 'Parch'], axis=1)
```

```python
data
```

```python
x = data[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked_Q', 'Embarked_S', 'FamilySize']]
y = data['Survived']
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
